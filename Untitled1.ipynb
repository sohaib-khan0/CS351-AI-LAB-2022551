```python
import networkx as nx
import matplotlib.pyplot as plt

# Example: Graph Coloring for University Exam Scheduling

# Create a graph to represent courses and their conflicts
def create_course_graph():
    G = nx.Graph()
    # Add courses as nodes
    courses = ["Math", "Physics", "Chemistry", "Biology", "History", "Art"]
    G.add_nodes_from(courses)

    # Add edges (conflicts between courses)
    conflicts = [("Math", "Physics"),
                 ("Math", "Chemistry"),
                 ("Physics", "Chemistry"),
                 ("Biology", "History"),
                 ("History", "Art"),
                 ("Chemistry", "Biology")]

    G.add_edges_from(conflicts)
    return G

# Function to visualize the graph
def visualize_graph(G, colors=None):
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_color=colors, font_weight='bold', node_size=800)
    plt.show()

# Backtracking algorithm for graph coloring
def is_valid(graph, node, color, colors):
    for neighbor in graph.neighbors(node):
        if colors[neighbor] == color:
            return False
    return True

def graph_coloring_util(graph, colors, node_index, num_colors):
    if node_index == len(graph.nodes()):
        return True

    node = list(graph.nodes())[node_index]
    for color in range(num_colors):
        if is_valid(graph, node, color, colors):
            colors[node] = color
            if graph_coloring_util(graph, colors, node_index + 1, num_colors):
                return True
            colors[node] = -1  # Backtrack

    return False

def graph_coloring(graph, num_colors):
    colors = {node: -1 for node in graph.nodes()}
    if graph_coloring_util(graph, colors, 0, num_colors):
        return colors
    else:
        print("Solution does not exist")
        return None

# Main function
def main():
    # Create the course graph
    course_graph = create_course_graph()

    # Visualize initial graph
    print("Initial course graph:")
    visualize_graph(course_graph)

    # Attempt to color the graph with a maximum of 3 colors (time slots)
    num_colors = 3
    colors = graph_coloring(course_graph, num_colors)

    if colors:
        print("Color assignments (time slots):")
        print(colors)

        # Prepare colors for visualization
        color_map = ['red', 'green', 'blue', 'orange']
        node_colors = [color_map[color] for color in colors.values()]

        # Visualize the colored graph
        print("Graph after coloring (time slots assigned):")
        visualize_graph(course_graph, colors=node_colors)

if __name__ == "__main__":
    main()
```
