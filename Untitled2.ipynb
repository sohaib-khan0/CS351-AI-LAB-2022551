```python
import random
from collections import deque
import math

# Non-AI Version: Player vs Computer
def player_vs_computer():
    number_to_guess = random.randint(1, 100)
    attempts = 0
    print("\n--- Player vs Computer ---")
    print("I have chosen a number between 1 and 100. Try to guess it!")

    while True:
        player_guess = int(input("Enter your guess: "))
        attempts += 1

        if player_guess < number_to_guess:
            print("Higher!")
        elif player_guess > number_to_guess:
            print("Lower!")
        else:
            print(f"Congratulations! You guessed the number {number_to_guess} in {attempts} attempts.")
            break

# AI Version with Binary Search: Computer vs Player
def ai_binary_search():
    number_to_guess = int(input("\n--- Computer vs Player (Binary Search) ---\nPlease enter a number between 1 and 100: "))
    low, high = 1, 100
    attempts = 0

    while True:
        computer_guess = (low + high) // 2
        attempts += 1
        print(f"Computer guesses: {computer_guess}")

        if computer_guess < number_to_guess:
            print("Higher!")
            low = computer_guess + 1
        elif computer_guess > number_to_guess:
            print("Lower!")
            high = computer_guess - 1
        else:
            print(f"Computer guessed the number {computer_guess} in {attempts} attempts.")
            break

# BFS Version
def ai_bfs():
    number_to_guess = int(input("\n--- AI using BFS ---\nPlease enter a number between 1 and 100: "))
    queue = deque(range(1, 101))
    attempts = 0

    while queue:
        computer_guess = queue.popleft()
        attempts += 1
        print(f"Computer guesses: {computer_guess}")

        if computer_guess < number_to_guess:
            print("Higher!")
            queue.extend(range(computer_guess + 1, 101))
        elif computer_guess > number_to_guess:
            print("Lower!")
        else:
            print(f"Computer guessed the number {computer_guess} in {attempts} attempts.")
            break

# DFS Version
def ai_dfs():
    number_to_guess = int(input("\n--- AI using DFS ---\nPlease enter a number between 1 and 100: "))
    stack = list(range(1, 101))
    attempts = 0

    while stack:
        computer_guess = stack.pop()
        attempts += 1
        print(f"Computer guesses: {computer_guess}")

        if computer_guess < number_to_guess:
            print("Higher!")
            stack.extend(range(computer_guess + 1, 101))
        elif computer_guess > number_to_guess:
            print("Lower!")
        else:
            print(f"Computer guessed the number {computer_guess} in {attempts} attempts.")
            break

# Simulated Annealing Version
def ai_simulated_annealing():
    number_to_guess = int(input("\n--- AI using Simulated Annealing ---\nPlease enter a number between 1 and 100: "))
    current_guess = random.randint(1, 100)
    attempts = 0
    temperature = 1000.0
    cooling_rate = 0.99

    while True:
        attempts += 1
        print(f"Computer guesses: {current_guess}")

        if current_guess < number_to_guess:
            print("Higher!")
            current_guess += 1
        elif current_guess > number_to_guess:
            print("Lower!")
            current_guess -= 1
        else:
            print(f"Computer guessed the number {current_guess} in {attempts} attempts.")
            break

        # Simulated annealing step
        if random.uniform(0, 1) < temperature:
            current_guess = random.randint(1, 100)

        temperature *= cooling_rate

# Main function to run the game
def main():
    while True:
        print("\nSelect Game Mode:")
        print("1. Player vs Computer")
        print("2. AI using Binary Search")
        print("3. AI using BFS")
        print("4. AI using DFS")
        print("5. AI using Simulated Annealing")
        print("6. Exit")

        choice = input("Enter your choice (1-6): ")

        if choice == '1':
            player_vs_computer()
        elif choice == '2':
            ai_binary_search()
        elif choice == '3':
            ai_bfs()
        elif choice == '4':
            ai_dfs()
        elif choice == '5':
            ai_simulated_annealing()
        elif choice == '6':
            print("Exiting the game.")
            break
        else:
            print("Invalid choice. Please select again.")

if __name__ == "__main__":
    main()
```
